```python
import numpy as np

positionData = np.arange(10, 110, 10)
oscillationData1 = np.array([1.211, 1.225, 1.238, 1.252, 1.268, 1.276, 1.289, 1.302, 1.315, 1.328])
oscillationData2 = np.array([1.134, 1.158, 1.186, 1.218, 1.242, 1.281, 1.316, 1.362, 1.405, 1.457])
```

```python
list(positionData)
```

```python
#import numpy as np
import matplotlib.pyplot as plt

f = plt.figure()

pol1 = np.polyfit(positionData, oscillationData1, 1)
pol_graph_1 = np.polyval(pol1, positionData)

pol2 = np.polyfit(positionData, oscillationData2, 1)
pol_graph_2 = np.polyval(pol2, positionData)

plt.plot(positionData, oscillationData1, 'k^')
plt.plot(positionData, oscillationData2, 'bo')

plt.plot(positionData, pol_graph_1, 'y--')
plt.plot(positionData, pol_graph_2, 'm--')

plt.title('Графік залежності Т від положення тягарця')
plt.ylabel('$T, с$')
plt.xlabel('$X, мм$')

#idx = np.argwhere(np.diff(np.sign(pol_graph_2 - pol_graph_1))).flatten()
#plt.plot(positionData[idx], pol_graph_2[idx], 'r*')
intersection_point = [53, 1.268]
plt.plot(intersection_point[0], intersection_point[1],'r*') 

plt.grid()
plt.show()
f.savefig("graph.pdf", bbox_inches='tight')
```
