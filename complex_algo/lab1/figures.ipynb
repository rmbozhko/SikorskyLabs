```python
data = []
with open('data_file.txt', 'r') as file:
    for line in file:
        data.append(line.rstrip())
data
```

```python
elem_num = []
timing = []
iterations_num = []
for elem in data:
    elems = elem.split(',')
    elem_num.append(int(elems[0]))
    timing.append(int(elems[1]))
    iterations_num.append(int(elems[2]))
elem_num
```

```python
timing
```

```python
iterations_num
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.plot(elem_num, timing, 'ro')
#np.arange(min(timing), max(timing), 50) - for recursion stats
plt.plot(elem_num, np.array(elem_num), 'b--')
plt.xlabel("Кількість елементів")
plt.ylabel("Кількість мікросекунд")
```

```python
import numpy
plt.xlabel("Кількість елементів")
plt.ylabel("Кількість ітерацій")
plt.plot(elem_num, iterations_num, 'ro')
plt.plot(elem_num, numpy.array(elem_num), 'b--')
```
